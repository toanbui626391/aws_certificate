### Spqrk process data with Watermark
Data Processing with Watermark:
- Early Enough Data (Before Watermark): Data with event times earlier than the current watermark is considered fully processed. Spark can finalize windowed aggregations for these events and potentially emit results (depending on the output mode).
- Late, But Within Threshold (Between Watermark and Watermark + Threshold): Data with event times falling within this range is considered late but still usable. Spark includes these events in the windowed aggregations, potentially updating earlier results.
- Very Late Data (Beyond Watermark + Threshold): Data exceeding the watermark + threshold is typically dropped. Since it's very unlikely to be relevant for the current window, dropping it avoids unnecessary processing and maintains state efficiency.
